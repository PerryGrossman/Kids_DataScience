Floor numbering is the numbering scheme used for a building's floors. There are two major schemes in use across the world. In one system, used in the majority of Europe, the ground floor is the floor on the ground and often has no number or is assigned the number zero. Therefore, the next floor up is assigned the number 1 and is the first floor.
The other system, used primarily in the United States and Canada, counts the bottom floor as number 1 or first floor.

```python
# Create a list of numbers
vals = [3, -8, 2, 7, 6, 2, 5, 12, 4, 9]
```

```python
#Find the even numbers
evens = []
for v in vals:
    if v%2 == 0:
        evens.append(v)
#How is this working?
```

```python
evens
```

```python
squares = []
for v in vals:
    squares.append(v*v)
```

```python
squares
```

```python
bigsquares = []
for v in vals:
    s = v*v
    if s > 10:
        bigsquares.append(s)
```
